```python
from scipy import stats
import numpy as np

chocolate = [25, 30, 28, 35, 32, 29, 33]
vanilla   = [20, 22, 25, 28, 24, 26, 23]

print("Average Sales:")
print(f"Chocolate: {np.mean(chocolate):.1f}")
print(f"Vanilla:   {np.mean(vanilla):.1f}")

t_stat, p_value = stats.ttest_ind(chocolate, vanilla)

print(f"\np-value = {p_value:.4f}")

if p_value < 0.05:
    print("Result: Chocolate sells SIGNIFICANTLY more!")
else:
    print("Result: No real difference (just luck).")
```
