```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# User input
x = sp.symbols('x')
f_expr = sp.sin(x) + x**2 / 2
f = sp.lambdify(x, f_expr, 'numpy')
df = sp.diff(f_expr, x)
df_func = sp.lambdify(x, df, 'numpy')

# Plot
X = np.linspace(-5, 5, 100)
plt.plot(X, f(X), label=f'f(x) = {f_expr}', color='blue')
plt.plot(X, df_func(X), label=f"f'(x) = {df}", color='red')
plt.legend()
plt.title('Function and Its Derivative')
plt.grid(True)
plt.show()
```
