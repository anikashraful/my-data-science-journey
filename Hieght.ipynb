```python
from scipy import stats
import numpy as np

boys  = [160, 165, 170, 168, 172]
girls = [155, 158, 160, 162, 159]

print("Average :")
print(f"Boys: {np.mean(boys):.1f}")
print(f"Girls:   {np.mean(girls):.1f}")

t_stat, p_value = stats.ttest_ind(boys, girls)

print(f"\np-value = {p_value:.4f}")

if p_value < 0.05:
    print("Result: Boys more!")
else:
    print("Result: No real difference (just luck).")
```
